```javascript
/**
 * @param {number[]} nums
 * @param {number} k
 * @return {number}
 */
var findKthLargest = function(nums, k) {
    // 解法1--先想到的就是快速排序
    // quickSort(nums,0,nums.length-1);
    // return nums[nums.length-k];


    // 解法2--堆排序，构造堆，一遍交换一遍构造，被交换的堆顶元素不参与构造了，直至有序数组。
    let heapSize=nums.length
    buildMaxHeap(nums,heapSize) // 构建好了一个大顶堆
        // 进行下沉 大顶堆是最大元素下沉到末尾
        for(let i=nums.length-1;i>=nums.length-k+1;i--){
            swap(nums,0,i)
            --heapSize // 下沉后的元素不参与到大顶堆的调整
            // 重新调整大顶堆
            maxHeapify(nums, 0, heapSize);
        }
        return nums[0]
   // 自下而上构建一颗大顶堆
   function buildMaxHeap(nums,heapSize){
     for(let i=Math.floor(heapSize/2)-1;i>=0;i--){
        maxHeapify(nums,i,heapSize)
     }
   }
   // 从左向右，自上而下的调整节点
   function maxHeapify(nums,i,heapSize){
       let l=i*2+1
       let r=i*2+2
       let largest=i
       if(l < heapSize && nums[l] > nums[largest]){
           largest=l
       }
       if(r < heapSize && nums[r] > nums[largest]){
           largest=r
       }
       if(largest!==i){
           swap(nums,i,largest) // 进行节点调整
           // 继续调整下面的非叶子节点
           maxHeapify(nums,largest,heapSize)
       }
   }
//    交换
   function swap(a,  i,  j){
        let temp = a[i];
        a[i] = a[j];
        a[j] = temp;
   }

};
//解法一：使用堆排序来解决这个问题——建立一个大顶堆，做k−1 次删除操作后,堆顶元素就是我们要找的答案


//解法二：使用快排将数组排好序之后，这样就可以拿到第k大的元素,不行
// function quickSort(nums,start,end){
//     if (start < end) {
//         //定义左指针和有指针，分别指向头和尾
//         let left = start
//         let right = end;
//         //定义一个排序基准,现已首为基准
//         let base = nums[start];
//         // 循环条件为 left < right
//         while(left < right){
//             //从右向左找，比base大，right--，是不需要动的
//             while(left < right && nums[right] >= base){
//                 right--;
//             }
//             //否则将right指向的值赋值给left指向的值
//             nums[left] = nums[right];
//             //从左向右找，比base小，left++
//             while(left < right && nums[left] < base){
//                 left++;
//             }
//             //否则将left指向的值赋值给right指向的值
//             nums[right] = nums[left];
//         }
//         //此时left == right，循环结束，使用base替换这个位置的数字，为什么要替换，因为....
//         nums[left] = base;
//         //分而治之
//         quickSort(nums,start,left-1);
//         quickSort(nums,left+1,end);
//     }
// }
```

## 42 接雨水

```javascript
//  思路单调递减栈,找到一个比当前高的柱子，弹出栈顶
var trap = function(height) {
    // 定义一个空栈
    let stack = [];
    let sum = 0;
    //遍历柱子的高度
    for(let i = 0 ;i < height.length ; i++){
        //满足：栈不为空，且当前元素值大于入栈下表对应的值=>弹出栈顶
        while(stack.length > 0 && height[i] > height[stack[stack.length - 1]]){
            let top = stack.pop();
            if(stack.length == 0)
                break;
            //形成凹槽计算面积
            let dis = i - stack[stack.length - 1] - 1;
            let area = Math.min(height[i],height[stack[stack.length - 1]]) - height[top];
            sum += dis * area;
        }
        stack.push(i);
    }
    return sum;
};
```

## 239 滑动窗口最大值

```javascript
var maxSlidingWindow = function(nums, k) {
    const n = nums.length;
    const prefixMax = new Array(n).fill(0);
    const suffixMax = new Array(n).fill(0);
    for (let i = 0; i < n; i++) {
        if (i % k === 0) {
            prefixMax[i] = nums[i];
        } else {
            prefixMax[i] = Math.max(prefixMax[i - 1], nums[i]);
        }
    }
    for (let i = n - 1; i >= 0; --i) {
        if (i === n || (i + 1) % k === 0) {
            suffixMax[i] = nums[i];
        } else {
            suffixMax[i] = Math.max(suffixMax[i + 1], nums[i]);
        }
    }
    const ans = [];
    for (let i = 0; i < n - k + 1; i++) {
        ans.push(Math.max(suffixMax[i], prefixMax[i + k - 1]));
    }
    return ans;
};
```

## 84 柱状图中最大的矩形

```javascript
var largestRectangleArea = function(heights) {
    let maxArea = 0;
	const stack = [0];
	heights.push(0);
	const n = heights.length;

	for (let i = 1; i < n; i++) {
		let top = stack.at(-1);
        // 情况一
		if (heights[top] < heights[i]) {
			stack.push(i);
		}
        // 情况二
		if (heights[top] === heights[i]) {
			stack.pop(); 
			stack.push(i);
		}
        // 情况三
		if (heights[top] > heights[i]) {
			while (stack.length > 0 && heights[top] > heights[i]) {
				const h = heights[stack.pop()];
				const left = stack.at(-1) ?? -1;
                console.log("left是"+left);
                console.log(i);
				const w = i - left - 1;
                // 最大面积
				maxArea = Math.max(maxArea, w * h);
				top = stack.at(-1);
			}
			stack.push(i);
		}
	}
	return maxArea;
};
```

## 207 课程表

```javascript
using namespace std;

class Solution {
public:
    bool canFinish(int numCourses, vector<vector<int>>& prerequisites) {
        // 加速优化
        if(prerequisites.size() > (numCourses-1)*numCourses/2) return false;

        vector<vector<int>> adj(numCourses);
        vector<int> indegree(numCourses, 0);

        // 创建邻接表
        for (const auto& prereq : prerequisites) {
            adj[prereq[1]].push_back(prereq[0]);
            indegree[prereq[0]]++;
        }
        
        // 广度优先检查有没有环
        queue<int> q;
        for (int i = 0; i < numCourses; i++) {
            if (indegree[i] == 0) q.push(i);
        }
        
        int coursesCompleted = 0;
        while (!q.empty()) {
            int course = q.front();
            q.pop();
            coursesCompleted++;
            
            for (int nextCourse : adj[course]) {
                indegree[nextCourse]--;
                if (indegree[nextCourse] == 0) q.push(nextCourse);
            }
        }
        
        return coursesCompleted == numCourses;
    }
};
```
